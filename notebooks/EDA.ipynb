```python
missing_percent = (df.isnull().sum() / len(df)) * 100

print(missing_percent.sort_values(ascending=False))
```

```python
df.drop(columns=['IsPEP', 'Occupation'], inplace=True)

print(df.shape)
```

```python
# Gender aur RiskCategoryId me mode (most frequent value) fill karenge

df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])

df['RiskCategoryId'] = df['RiskCategoryId'].fillna(
    df['RiskCategoryId'].mode()[0]
)

print(df[['Gender', 'RiskCategoryId']].isnull().sum())
```

```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])

df['AccountOpenDate'] = pd.to_datetime(df['AccountOpenDate'])

df['DateofBirth'] = pd.to_datetime(
    df['DateofBirth'],
    errors='coerce'
)

print(df[['TransactionDate',
          'AccountOpenDate',
          'DateofBirth']].dtypes)
```

```python
median_dob = df['DateofBirth'].median()

df['DateofBirth'] = df['DateofBirth'].fillna(median_dob)

print(df['DateofBirth'].isnull().sum())
```

```python
from datetime import datetime

today = pd.Timestamp.today()

df['Age'] = (
    (today - df['DateofBirth']).dt.days // 365
)

print(df['Age'].describe())
```
